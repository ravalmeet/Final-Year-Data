```python
import pandas as pd
```

```python
bengali = pd.read_csv('final_bengali.csv')
gujarati = pd.read_csv('final_gujarati.csv')
hindi = pd.read_csv('final_hindi.csv')
marathi = pd.read_csv('final_marathi.csv')
punjabi = pd.read_csv('final_punjabi.csv')
tamil = pd.read_csv('final_tamil.csv')
telugu = pd.read_csv('final_telugu (1).csv')
```

```python
len_bengali = len(bengali)
len_gujarati = len(gujarati)
len_hindi = len(hindi)
len_marathi = len(marathi)
len_punjabi = len(punjabi)
len_tamil = len(tamil)
len_telugu = len(telugu)
```

```python
bengali_1 = bengali[bengali['result'] == 1]
bengali_0 = bengali[bengali['result'] == 0]

gujarati_1 = gujarati[gujarati['result'] == 1]
gujarati_0 = gujarati[gujarati['result'] == 0]

hindi_1 = hindi[hindi['result'] == 1]
hindi_0 = hindi[hindi['result'] == 0]

marathi_1 = marathi[marathi['result'] == 1]
marathi_0 = marathi[marathi['result'] == 0]

punjabi_1 = punjabi[punjabi['result'] == 1]
punjabi_0 = punjabi[punjabi['result'] == 0]

tamil_1 = tamil[tamil['result'] == 1]
tamil_0 = tamil[tamil['result'] == 0]

telugu_1 = telugu[telugu['result'] == 1]
telugu_0 = telugu[telugu['result'] == 0]
```

```python
random_bengali_1 = bengali_1[:2250]
random_bengali_0 = bengali_0[:2250]

random_gujarati_1 = gujarati_1[:2250]
random_gujarati_0 = gujarati_0[:2250]

random_hindi_1 = hindi_1[:2413]
random_hindi_0 = hindi_0[:2250]

random_marathi_1 = marathi_1[:2250]
random_marathi_0 = marathi_0[:2250]

random_punjabi_1 = punjabi_1[:2250]
random_punjabi_0 = punjabi_0[:2250]

random_tamil_1 = tamil_1[:2452]
random_tamil_0 = tamil_0[:2250]

random_telugu_1 = telugu_1[:2250]
random_telugu_0 = telugu_0[:2250]
```

```python
combined_bengali = pd.concat([random_bengali_0, random_bengali_1], ignore_index=True)
combined_gujarati = pd.concat([random_gujarati_0, random_gujarati_1], ignore_index=True)
combined_hindi = pd.concat([random_hindi_0, random_hindi_1], ignore_index=True)
combined_marathi = pd.concat([random_marathi_0, random_marathi_1], ignore_index=True)
combined_punjabi = pd.concat([random_punjabi_0, random_punjabi_1], ignore_index=True)
combined_tamil = pd.concat([random_tamil_0, random_tamil_1], ignore_index=True)
combined_telugu = pd.concat([random_telugu_0, random_telugu_1], ignore_index=True)
```

```python
print("Length of Bengali DataFrame:", len(combined_bengali))
print("Length of Gujarati DataFrame:", len(combined_gujarati))
print("Length of Hindi DataFrame:", len(combined_hindi))
print("Length of Marathi DataFrame:", len(combined_marathi))
print("Length of Punjabi DataFrame:", len(combined_punjabi))
print("Length of Tamil DataFrame:", len(combined_tamil))
print("Length of Telugu DataFrame:", len(combined_telugu))
```

```python
combined_bengali.rename(columns={'bengali_text': 'text'}, inplace=True)
combined_gujarati.rename(columns={'guj_text': 'text'}, inplace=True)
combined_hindi.rename(columns={'Text': 'text'}, inplace=True)
combined_marathi.rename(columns={'marathi_text': 'text'}, inplace=True)
combined_punjabi.rename(columns={'punjabi_text': 'text'}, inplace=True)
combined_tamil.rename(columns={'tamil_text': 'text'}, inplace=True)
combined_telugu.rename(columns={'telugu_text': 'text'}, inplace=True)
```

```python
all_combined = pd.concat([combined_bengali, combined_gujarati, combined_hindi, combined_marathi, combined_punjabi, combined_tamil, combined_telugu], ignore_index=True)
```
